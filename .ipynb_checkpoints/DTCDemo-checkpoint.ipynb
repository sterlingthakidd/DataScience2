```python
#1. Set root condition to be <= 2.5 cm petal length cm
#2. If less than 2.5, classify as 'setosa'
# 3. Use gini impurity loss calculation to systematically check the impurity
    # of the 2nd node.
    # Sort dataframe by petal length. Then, take average of adjacent values.
    # Use average as boundary check.
    # For each leaf, 1 - (probability of yes)^2 - (probability of no)^2
    # For entire tree, weight the impurity of each leaf
        # (yes + no of leaf / total observations across tree) * gini of that leaf
        # Sum all of these weight impuritys together for the entire tree
```

```python
def getNextBoundary(i, sortedDF):
    if i < (len(sortedDF)-1):
        first = sortedDF.iloc[i]["PetalLengthCm"]
        second = sortedDF.iloc[i+1]["PetalLengthCm"]
        return (first + second) / 2
    return (sortedDF.iloc[i-1]["PetalLengthCm"] + sortedDF.iloc[i]["PetalLengthCm"]/2)
```

```python
# {1: [['yes setosa', 'no setosa'], ['yes virginica', 'no virgina'], ['yes versicolor', 'no versicolor']])
def trainTree(df):
    trees = {}
    setosa = [0, 0] # yes, no
    virgina = [0, 0] # yes, no
    versicolor = [0, 0] # yes, no
    treeNumber = 1
    trees[treeNumber] = [setosa, versicolor, virgina]
    for i in range(len(df)):
    # Root node check
        if df.iloc[i]["PetalLengthCm"] <= 2.5:
            if df.iloc[i]["Species"] == 'Iris-setosa':
                trees[treeNumber][0][0] = trees[treeNumber][0][0] + 1
            else:
                trees[treeNumber][0][1] = trees[treeNumber][0][1] + 1
        else:
            # second node check
            boundary = getNextBoundary(i, df)
            if df.iloc[i]["PetalLengthCm"] <= boundary:
                if df.iloc[i]["Species"] == 'Iris-versicolor':
                    trees[treeNumber][1][0] = trees[treeNumber][1][0] + 1
                else:
                    trees[treeNumber][1][1] = trees[treeNumber][1][1] + 1
            else:
                if df.iloc[i]["Species"] == 'Iris-virginica':
                    trees[treeNumber][2][0] = trees[treeNumber][2][0] + 1
                else:
                    trees[treeNumber][2][1] = trees[treeNumber][2][1] + 1
    return trees
    #treeNumber = treeNumber + 1
```

```python
x = trainTree(df)
x
```

```python
def giniImpurityCalc(tree, key):
    denomSetosa = tree[key][0][0] + tree[key][0][1]
    leaf1 = 1 - (tree[key][0][0]/denomSetosa)**2 - (tree[key][0][1]/denomSetosa)**2
    denomVersi = tree[key][1][0] + tree[key][1][1]
    leaf2 = 1 - (tree[key][1][0]/denomVersi)**2 - (tree[key][1][1]/denomVersi)**2
    denomVirg = tree[key][2][0]+ tree[key][2][1]
    leaf3 = 1 - (tree[key][2][0]/denomVirg)**2 - (tree[key][2][1]/denomVirg)**2
    
    leaf1Obs = tree[key][0][0] + tree[key][0][1]
    leaf2Obs = tree[key][1][0] + tree[key][1][1]
    leaf3Obs = tree[key][2][0] + tree[key][2][1]
    totalObs = tree[key][0][0] + tree[key][0][1] + tree[key][1][0] + tree[key][1][1] + tree[key][2][0] + tree[key][2][1]
    
    giniScoree = leaf1 * (leaf1Obs/totalObs) + leaf2 * (leaf2Obs/totalObs) + leaf3 * (leaf3Obs/totalObs)
    return giniScoree
```

```python
gi = giniImpurityCalc(x, 1)
```
