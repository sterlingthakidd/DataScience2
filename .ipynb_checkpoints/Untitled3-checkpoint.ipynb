```python
def getNextBoundary(i, sortedDF):
    if i < (len(sortedDF)-1):
        first = sortedDF.iloc[i]["Sample Number"]
        second = sortedDF.iloc[i+1]["Sample Number"]
        return (first + second) / 2
    return (sortedDF.iloc[i-1]["Sample Number"] + sortedDF.iloc[i]["Sample Number"]/2)
```

```python
# {1: [['yes setosa', 'no setosa'], ['yes virginica', 'no virgina'], ['yes versicolor', 'no versicolor']])
def trainTree(df):
    trees = {}
    adeliae = [0, 0] # yes, no
    antarctica = [0, 0] # yes, no
    papua = [0, 0] # yes, no
    treeNumber = 1
    trees[treeNumber] = [adeliae, antarctica, papua]
    for i in range(len(df)):
    # Root node check
        if df.iloc[i]["Sample Number"] <= 2.5:
            if df.iloc[i]["Species"] == 'Adelie Penguin (Pygoscelis adeliae)':
                trees[treeNumber][0][0] = trees[treeNumber][0][0] + 1
            else:
                trees[treeNumber][0][1] = trees[treeNumber][0][1] + 1
        else:
            # second node check
            boundary = getNextBoundary(i, df)
            if df.iloc[i]["Sample Number"] <= boundary:
                if df.iloc[i]["Species"] == 'Gentoo penguin (Pygoscelis papua)':
                    trees[treeNumber][1][0] = trees[treeNumber][1][0] + 1
                else:
                    trees[treeNumber][1][1] = trees[treeNumber][1][1] + 1
            else:
                if df.iloc[i]["Species"] == 'Chinstrap penguin (Pygoscelis antarctica)':
                    trees[treeNumber][2][0] = trees[treeNumber][2][0] + 1
                else:
                    trees[treeNumber][2][1] = trees[treeNumber][2][1] + 1
    return trees
    #treeNumber = treeNumber + 1
```

```python
x = trainTree(df)
x
```

```python
def giniImpurityCalc(tree, key):
    denomAdeliae = tree[key][0][0] + tree[key][0][1]
    leaf1 = 1 - (tree[key][0][0]/denomAdeliae)**2 - (tree[key][0][1]/denomAdeliae)**2
    denomPapua = tree[key][1][0] + tree[key][1][1]
    leaf2 = 1 - (tree[key][1][0]/denomPapua)**2 - (tree[key][1][1]/denomPapua)**2
    denomAntarctica = tree[key][2][0]+ tree[key][2][1]
    leaf3 = 1 - (tree[key][2][0]/denomAntarctica)**2 - (tree[key][2][1]/denomAntarctica)**2
    
    leaf1Obs = tree[key][0][0] + tree[key][0][1]
    leaf2Obs = tree[key][1][0] + tree[key][1][1]
    leaf3Obs = tree[key][2][0] + tree[key][2][1]
    totalObs = tree[key][0][0] + tree[key][0][1] + tree[key][1][0] + tree[key][1][1] + tree[key][2][0] + tree[key][2][1]
    
    giniScoree = leaf1 * (leaf1Obs/totalObs) + leaf2 * (leaf2Obs/totalObs) + leaf3 * (leaf3Obs/totalObs)
    return giniScoree
```

```python
gi = giniImpurityCalc(x, 1)
```
