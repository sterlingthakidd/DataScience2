```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("train.csv")
df.head()
```

```python
sns.scatterplot(data=df, x="popularity", y="budget", hue="revenue")
```

```python
# Split into trian and test datasets
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
train_df.info()
```

```python
revenues = train_df[train_df.revenue == 'yes']
nonrevenue = train_df[train_df.revenue == "no"]
```

```python
# Train model
def trainModel(trainDF, x, y):
    slopes = []
    for i in range(len(trainDF)):
        slope = (trainDF.iloc[i]["budget"] - y) / (trainDF.iloc[i]["popularity"] - x)
        slopes.append(slope)
    
    return sum(slopes)/len(slopes)
b = 10000
m = trainModel( revenue, 0, b)
print("y = " + str(m) + "x + " + str(b))
```
